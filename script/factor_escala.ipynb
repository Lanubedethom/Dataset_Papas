```python
import cv2
import matplotlib.pyplot as plt

imagen = cv2.imread("../papas/6/1.jpg")
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

factor_escala = 0.005999880002399953

plt.imshow(gris, cmap='gray')
plt.title("Imagen con escala (en escala de grises)")

# Ajustar los ejes para mostrar la escala en centímetros
height, width = gris.shape
xticks = range(0, width, int(1/factor_escala))
yticks = range(0, height, int(1/factor_escala))
xlabels = range(0, len(xticks))
ylabels = range(0, len(yticks))

plt.xticks(ticks=xticks, labels=xlabels)
plt.yticks(ticks=yticks, labels=ylabels)

plt.xlabel("Centímetros")
plt.ylabel("Centímetros")
plt.axis("on")  
plt.show()
```
